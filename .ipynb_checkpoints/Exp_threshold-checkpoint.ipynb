### Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display(name,img):
    cv2.imshow(name,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
def plot(name,img):
    plt.axis("off")
    plt.imshow(img,cmap="gray")
    plt.title(name)
```

### Read the image and Convert to gray scale

```python
img = cv2.imread("gojo.png",1)
display("Gojo",img)
plot("Gojo",img)
```

```python
img_gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display("Gojo_Gray",img_gray)
plot("Gojo_Gray",img_gray)
```

### Global Thresholding

#### Binary

```python
ret,t_b   = cv2.threshold(img_gray,200,300,cv2.THRESH_BINARY)

#display("Gojo_Binary_Threshhold",t_b)
plot("Gojo_Binary_Threshhold",t_b)
```

```python
ret,t_b_i = cv2.threshold(img_gray,200,300,cv2.THRESH_BINARY_INV)

#display("Gojo_Binary_Inverse_Threshhold",t_b_i)
plot("Gojo_Binary_Inverse_Threshhold",t_b_i)
```

#### TRUNIC

```python
ret,t_t = cv2.threshold(img_gray,200,300,cv2.THRESH_TRUNC)
plot("Gojo_Trunic",t_t)
```

#### To Zero

```python
ret,t_tz =cv2.threshold(img_gray,86,255,cv2.THRESH_TOZERO)
plot("Gojo_ToZero",t_tz)
```

```python
ret,t_tz_i =cv2.threshold(img_gray,86,255,cv2.THRESH_TOZERO_INV)
plot("Gojo_ToZero_Inv",t_tz_i)
```
